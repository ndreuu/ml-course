## Свойства [датасета](https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/) 
`MSSubClass`: Определяет тип жилья, участвующего в продаже.

`MSZoning`: Определяет общую классификацию зонирования объекта продажи.

`LotFrontage`: Линейные метры улицы, соединенной с объектом недвижимости

`LotArea`: Размер участка в квадратных футах

`Street`: Тип подъездной дороги к объекту недвижимости

`Alley`: Тип подъездной аллеи к собственности

`LotShape`: Общая форма собственности

`LandContour`: Ровность участка

`Utilities`: Тип доступных утилит

`LotConfig`: Конфигурация партии

`LandSlope`: Уклон участка

`Neighborhood`: Физические местоположения в пределах города Эймс

`Condition1`: Близость к различным условиям

`Condition2`: Близость к различным условиям (если присутствует более одного)

`BldgType`: Тип жилого помещения

`HouseStyle`: Стиль жилища

`OverallQual`: Оценивает общий материал и отделку дома

`OverallCond`: Оценивает общее состояние дома

`YearBuilt`: Первоначальная дата постройки

`YearRemodAdd`: Дата реконструкции (совпадает с датой строительства, если реконструкция или дополнения отсутствуют)

`RoofStyle`: Тип кровли

`RoofMatl`: Кровельный материал

`Exterior1st`: Наружное покрытие дома

`Exterior2nd`: Наружное покрытие дома (если используется более одного материала)

`MasVnrType`: Тип облицовки каменной кладки

`MasVnrArea`: Площадь облицовки каменной кладки в квадратных футах

`ExterQual`: Оценивает качество материала, из которого выполнена внешняя отделка

`ExterCond`: Оценивает текущее состояние материала снаружи

`Foundation`: Тип фундамента

`BsmtQual`: Оценивает высоту подвала

`BsmtCond`: Оценивает общее состояние подвала

`BsmtExposure`: Относится к стенам на уровне пешеходной дорожки или сада

`BsmtFinType1`: Оценка готовой площади подвала

`BsmtFinSF1`: Тип 1 готовые квадратные футы

`BsmtFinType2`: Оценка готовой площади подвала (при наличии нескольких типов)

`BsmtFinSF2`: Тип 2 готовые квадратные футы

`BsmtUnfSF`: Недостроенные квадратные метры подвальной площади

`TotalBsmtSF`: Общая площадь подвала в квадратных футах

`Heating`: Тип отопления

`HeatingQC`: Качество и состояние отопления

`CentralAir`: Центральное кондиционирование воздуха

`Electrical`: Электрическая система

`1stFlrSF`: Площадь первого этажа в квадратных футах

`2ndFlrSF`: Площадь второго этажа в квадратных футах


`LowQualFinSF`: Квадратные метры с отделкой низкого качества (все этажи)

`GrLivArea`: Жилая площадь надземной части (цокольный этаж), кв. фут

`BsmtFullBath`: Цокольный этаж с полностью оборудованными ванными комнатами

`BsmtHalfBath`: Половина ванных комнат на цокольном этаже

`FullBath`: Полностью оборудованные ванные комнаты высшего класса

`HalfBath`: Половина ванн высшего сорта

`Bedroom`: Спальни выше этажом (не включая спальни на цокольном этаже)

`Kitchen`: Кухни высшего класса

`KitchenQual`: Качество кухни

`TotRmsAbvGrd`: Общее количество номеров выше класса (без учета ванных комнат)

`Functional`: Функциональность дома (предполагается типичной, если только не требуются вычеты)

`Fireplaces`: Количество каминов

`GarageCond`: Состояние гаража

`MiscVal`: Значения различных функций

`MoSold`: Проданный месяц (ММ)

## Заполняем пропущенные значения

```python

train_labels = train['SalePrice'].reset_index(drop=True)
train_features = train.drop(['SalePrice'], axis=1)
test_features = test

all_features = pd.concat([train_features, test_features]).reset_index(drop=True)
all_features.shape

def percent_missing(df):
    data = pd.DataFrame(df)
    df_cols = list(pd.DataFrame(data))
    dict_x = {}
    for i in range(0, len(df_cols)):
        dict_x.update({df_cols[i]: round(data[df_cols[i]].isnull().mean()*100,2)})
    
    return dict_x

missing = percent_missing(all_features)
df_miss = sorted(missing.items(), key=lambda x: x[1], reverse=True)
print('Percent of missing data')
df_miss[0:10]
```

```python
# Визуализация пропущенных значений
sns.set_style("white")
f, ax = plt.subplots(figsize=(8, 7))
sns.set_color_codes(palette='deep')
missing = round(train.isnull().mean()*100,2)
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar(color="b")
ax.xaxis.grid(False)
ax.set(ylabel="Percent of missing values")
ax.set(xlabel="Features")
ax.set(title="Percent missing data by feature")
sns.despine(trim=True, left=True)
```

Теперь мы можем просмотреть каждое из вышеперечисленных свойств и вписать недостающие значения для каждой из них.

```python
# Некоторые нечисловые предикторы хранятся в виде чисел; преобразовываем их в строки 
all_features['MSSubClass'] = all_features['MSSubClass'].apply(str)
all_features['YrSold'] = all_features['YrSold'].astype(str)
all_features['MoSold'] = all_features['MoSold'].astype(str)
```

```python
def handle_missing(features):
    features['Functional'] = features['Functional'].fillna('Typ')
    features['Electrical'] = features['Electrical'].fillna("SBrkr")
    features['KitchenQual'] = features['KitchenQual'].fillna("TA")
    features['Exterior1st'] = features['Exterior1st'].fillna(features['Exterior1st'].mode()[0])
    features['Exterior2nd'] = features['Exterior2nd'].fillna(features['Exterior2nd'].mode()[0])
    features['SaleType'] = features['SaleType'].fillna(features['SaleType'].mode()[0])
    features['MSZoning'] = features.groupby('MSSubClass')['MSZoning'].transform(lambda x: x.fillna(x.mode()[0]))
    features["PoolQC"] = features["PoolQC"].fillna("None")
    for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
        features[col] = features[col].fillna(0)
    for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
        features[col] = features[col].fillna('None')
    for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
        features[col] = features[col].fillna('None')
        
    features['LotFrontage'] = features.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

    objects = []
    for i in features.columns:
        if features[i].dtype == object:
            objects.append(i)
    features.update(features[objects].fillna('None'))
        
    numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    numeric = []
    for i in features.columns:
        if features[i].dtype in numeric_dtypes:
            numeric.append(i)
    features.update(features[numeric].fillna(0))    
    return features

all_features = handle_missing(all_features)
```

```python
missing = percent_missing(all_features)
df_miss = sorted(missing.items(), key=lambda x: x[1], reverse=True)
print('Percent of missing data')
df_miss[0:10]
```
